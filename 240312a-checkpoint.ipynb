```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import param
import panel as pn

# 3/12/2024
# Sijie Hu

# 1 EPS df
# source: https://www.macrotrends.net/stocks/charts/MSFT/microsoft/pe-ratio
dfepsnvda = pd.read_csv('NVDA PE Ratio Historical Data.csv')
dfepsnvda['Date'] = pd.to_datetime(dfepsnvda['Date'])

dfepsmsft = pd.read_csv('MSFT PE Ratio Historical Data.csv')
dfepsmsft['Date'] = pd.to_datetime(dfepsmsft['Date'])

dfepsgoog = pd.read_csv('GOOG PE Ratio Historical Data.csv')
dfepsgoog['Date'] = pd.to_datetime(dfepsgoog['Date'])

NVDA_eps = dfepsnvda.copy()
MSFT_eps = dfepsmsft.copy()
GOOG_eps = dfepsgoog.copy()

# 2 Price nvda
dfnvda = pd.read_csv('NVDA.csv')
dfnvda['Date'] = pd.to_datetime(dfnvda['Date'])

# 2 Price others
dfspy = pd.read_csv('SPY.csv')
dfxlk = pd.read_csv('XLK.csv')
dfmsft = pd.read_csv('MSFT.csv')
dfspy['Date'] = pd.to_datetime(dfspy['Date'])
dfxlk['Date'] = pd.to_datetime(dfxlk['Date'])
dfmsft['Date'] = pd.to_datetime(dfmsft['Date'])

NVDA_price = dfnvda.copy()
SPY_price = dfspy.copy()
XLK_price = dfxlk.copy()
MSFT_price = dfmsft.copy()

# 3 SP500 info
# dfsp500 = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
# dfsp500['Date added'] = pd.to_datetime(dfsp500['Date added'])
# dfsp500['Founded'] = dfsp500['Founded'].str[0:4]
# dfsp500['Founded'] = pd.to_datetime(dfsp500['Founded']).dt.year
# dfsp500 = dfsp500.drop(columns='CIK')

# dfsp500.to_csv('SP500.csv', index=False)
# COMP = dfsp500[dfsp500['Symbol'].isin(['NVDA','MSFT','AAPL','GOOG'])].copy()

dfsp500 = pd.read_csv('SP500.csv', parse_dates=['Date added'])
COMP = dfsp500[dfsp500['Symbol'].isin(['NVDA','MSFT','AAPL','GOOG'])].copy()
```

```python
# start charting
# when did NVDA enter SP500 compared to others?
fig, ax = plt.subplots(figsize=(12,6))

dfsp500.hist(ax=ax, column='Date added', bins=40)
ax.set_title('SP500 Entry Year Histogram')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Companies')
ax.text(COMP[COMP['Symbol']=='NVDA']['Date added'], 5, '<-NVDA')
ax.text(COMP[COMP['Symbol']=='MSFT']['Date added'], 5, '<-MSFT')
ax.text(COMP[COMP['Symbol']=='AAPL']['Date added'], 5, '<-AAPL')
ax.text(COMP[COMP['Symbol']=='GOOG']['Date added'], 5, '<-GOOG')
```

```python
# how was NVDA price compared to SP500?

# first we need to scale dfs to the same value in the beginning
# 1/2/2020

X_spy = NVDA_price.loc[0]['Close'] / SPY_price.loc[0]['Close']
SPY_price['Close multiplied'] = SPY_price['Close'] * X_spy

fig, ax = plt.subplots(figsize=(12,6))

NVDA_price.plot(ax=ax, x='Date', y='Close', label='NVDA')
SPY_price.plot(ax=ax, x='Date', y='Close multiplied', label='SPY')

ax.set_title('NVDA Performance in 2020 Compared to SPY')
ax.set_xlabel('Month')
ax.set_ylabel('Price')
ax.set_xlim(pd.Timestamp('2020-01-01'), pd.Timestamp('2020-12-31'))
ax.set_ylim(30,160)
ax.text(pd.Timestamp('2020-02-20'), 80, '<-Major selloff due to COVID')
```

```python
# how was NVDA price compared to Information Technology sector?

# first we need to scale dfs to the same value in the beginning
# 1/2/2020

X_xlk = NVDA_price.loc[0]['Close'] / XLK_price.loc[0]['Close']
XLK_price['Close multiplied'] = XLK_price['Close'] * X_xlk

fig, ax = plt.subplots(figsize=(12,6))

NVDA_price.plot(ax=ax, x='Date', y='Close', label='NVDA')
XLK_price.plot(ax=ax, x='Date', y='Close multiplied', label='XLK')
SPY_price.plot(ax=ax, x='Date', y='Close multiplied', label='SPY')

ax.set_title('NVDA Performance in 2020 Compared to XLK and SPY')
ax.set_xlabel('Month')
ax.set_ylabel('Price')
ax.set_xlim(pd.Timestamp('2020-01-01'), pd.Timestamp('2020-12-31'))
ax.set_ylim(30,160)
ax.text(pd.Timestamp('2020-02-20'), 80, '<-Major selloff due to COVID')

# what is the return of the year?
NVDA_return = NVDA_price.loc[252]['Close'] / NVDA_price.loc[0]['Close'] * 100
SPY_return = SPY_price.loc[252]['Close'] / SPY_price.loc[0]['Close'] * 100
XLK_return = XLK_price.loc[252]['Close'] / XLK_price.loc[0]['Close'] * 100

ax.text(pd.Timestamp('2020-11-15'), 120, f'{NVDA_return:.2f}% YTD')
ax.text(pd.Timestamp('2020-11-15'), 60, f'{SPY_return:.2f}% YTD')
ax.text(pd.Timestamp('2020-11-15'), 90, f'{XLK_return:.2f}% YTD')
```

```python
# interactive
import altair as alt
import plotly.graph_objs as go
import hvplot.pandas
import holoviews as hv
import datetime

pn.extension('vega', 'plotly')

# plt.rcParams.update(plt.rcParamsDefault)
# plt.rcParams.keys()
# https://stackoverflow.com/questions/12444716/how-do-i-set-the-figure-title-and-axes-labels-font-size

# plt.style.use('seaborn-v0_8')
# params = {'font.size': 8,
#           'axes.labelsize': 8,
#           'axes.titlesize': 8,
#           'xtick.labelsize': 8,
#           'ytick.labelsize': 8}
# plt.rcParams.update(params)

# class
class SJH(param.Parameterized):
    end_select = param.Selector(default=pd.Timestamp('2020-06-01'),
                                objects=[pd.Timestamp('2020-06-01'),
                                         pd.Timestamp('2021-06-01'),
                                         pd.Timestamp('2023-06-01'),
                                         pd.Timestamp('2024-03-01')],
                                label='End Year/Month')
    end = param.CalendarDate(datetime.date(2020, 6, 1),
                             label='End Year/Month')
    bins = param.Integer(default=20,
                         bounds=(10, 40),
                         step=5,
                         label='Histogram Bins')
    show_legend = param.Boolean(default=True,
                               label='Legend')

    def p1(self):
        # when did NVDA enter SP500 compared to others?
        
        a = dfsp500.hvplot.hist('Date added',
                                bins=self.bins)
        b = hv.Text(COMP[COMP['Symbol']=='NVDA']['Date added'].values[0],
                    50,
                    f"<- {str(COMP[COMP['Symbol']=='NVDA']['Date added'].values[0])[:10]}: NVDA entry",
                    halign='left')
        
        # COMP[COMP['Symbol']=='NVDA']['Date added'].dt.strftime('%Y-%m-%d').values[0]

        overlay = a*b
        overlay.opts(title='SP500 Entry Year Histogram vs NVDA (Source:Wiki)',
                     ylabel='Number of Companies',
                     show_legend=self.show_legend,
                     width=1600,
                     shared_axes=False)
        
        return overlay
        
    def p2(self):
        # how was NVDA price compared to SP500?
        
        a = NVDA_price.hvplot(x='Date',
                              y='Close',
                              label='NVDA')
        b = SPY_price.hvplot(x='Date',
                             y='Close multiplied',
                             label='SPY')
        c = XLK_price.hvplot(x='Date',
                             y='Close multiplied',
                             label='XLK')
        d = hv.Text(pd.Timestamp('2020-02-20'),
                    80,
                    '<-2/20/2020: Major selloff due to COVID',
                    halign='left')

        overlay1 = a*b*d
        overlay2 = a*b*c*d

        xlim = (pd.Timestamp('2020-01-01'),
                self.end_select)
        ylim = (30,
                NVDA_price[NVDA_price['Date']==self.end_select]['Close'].values[0]+30)
        
        overlay1.opts(title='NVDA Performance Compared to SP500 Industry (Source:YFinance)',
                      show_legend=self.show_legend,
                      xlim=xlim,
                      ylim=ylim,
                      width=800,
                      height=400)
        overlay2.opts(title='NVDA Performance Compared to Information Technology Sector (Source:YFinance)',
                      show_legend=self.show_legend,
                      xlim=xlim,
                      ylim=ylim,
                      width=800,
                      height=400)
        
        return overlay1 + overlay2
        
    def p3(self):
        # how was NVDA price compared to SP500?
        
        a = NVDA_price.hvplot(x='Date',
                              y='Close',
                              label='NVDA')
        b = SPY_price.hvplot(x='Date',
                             y='Close multiplied',
                             label='SPY')
        c = XLK_price.hvplot(x='Date',
                             y='Close multiplied',
                             label='XLK')
        d = hv.Text(pd.Timestamp('2020-02-20'),
                    100,
                    '<-2/20/2020: Major selloff due to COVID',
                    halign='left')
        ret = NVDA_price[NVDA_price['Date']=='2021-12-31']['Close'].values[0] / \
              NVDA_price[NVDA_price['Date']=='2020-01-02']['Close'].values[0] * 100
        
        e = hv.Text(pd.Timestamp('2021-12-31'),
                    380,
                    f'<- NVDA Return since 1/1/2020: {ret:.2f}%',
                    halign='left')

        overlay = a*b*c*d*e
        
        overlay.opts(title='NVDA Return % Compared to SP500 and Sector (Source:YFinance)',
                     show_legend=self.show_legend,
                     shared_axes=False,
                     width=1600,
                     height=300)
        
        return overlay

    def p4(self):
        # how is P/E compared to MSFT
        
        a = NVDA_eps.hvplot(x='Date', y='PE Ratio', label='NVDA', color='orange')
        b = MSFT_eps.hvplot(x='Date', y='PE Ratio', label='MSFT', color='green')
        c = GOOG_eps.hvplot(x='Date', y='PE Ratio', label='GOOG', color='purple')

        overlay1 = a*b
        overlay2 = a*c

        overlay1.opts(title='NVDA PE Ratio vs MSFT 2010-Present (Source:Macrotrends)',
                     show_legend=self.show_legend,
                     width=800,
                     height=400)
        
        overlay2.opts(title='NVDA PE Ratio vs GOOG 2010-Present (Source:Macrotrends)',
                     show_legend=self.show_legend,
                     width=800,
                     height=400)

        d = overlay1 + overlay2
        d.opts(shared_axes=False)

        return d

    def p5(self):
        # other comparison

        a = FINANCIAL.drop(columns=['marketCap',
                                    'trailingPE',
                                    'forwardPE'])
        
        a1 = a.hvplot.bar(x='symbol',
                          title='NVDA Current Stock Price vs MSFT GOOG (Source:YFinance)',
                          xlabel='Symbol',
                          ylabel='Price',
                          rot=90,
                          width=800,
                          height=500)
        
        b = FINANCIAL.drop(columns=['marketCap',
                                    'currentPrice',
                                    'fiftyTwoWeekLow',
                                    'fiftyTwoWeekHigh'])
        
        b1 = b.hvplot.bar(x='symbol',
                          title='NVDA Current P/E vs MSFT GOOG (Source:YFinance)',
                          xlabel='Symbol',
                          ylabel='P/E',
                          rot=90,
                          width=800,
                          height=500)

        a1.opts(shared_axes=False,
                show_legend=self.show_legend,
                legend_position='top_right',
                multi_level=False)

        b1.opts(shared_axes=False,
                show_legend=self.show_legend,
                legend_position='top_right',
                multi_level=False)

        return a1+b1

    def p6(self):
        # other comparison
        
        a = FINANCIAL[['symbol','marketCap']]
        
        a1 = a.hvplot.barh(x='symbol',
                          title='NVDA Market Cap vs MSFT GOOG (Source:YFinance)',
                          xlabel='Symbol',
                          ylabel='Trillion',
                          color='teal',
                          legend=self.show_legend,
                          bar_width=0.5,
                          width=1600,
                          height=300)
        
        return a1

sjh = SJH()
my_app = pn.Column(sjh.param, sjh.p1, sjh.p2, sjh.p4, sjh.p5, sjh.p6)
my_app.show()
```
